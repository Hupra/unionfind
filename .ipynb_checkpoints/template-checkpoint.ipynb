```python
%autoreload
from quickfind import *
from quickunion import *
from weightedquickunion import*
import time

with open("mediumUF.txt", "r+") as f:
    lines = f.readlines()
    n = int(lines[0])
    lines = lines[1:]
```

### Simple Quickfind

```python
start = time.time()

qf = QuickFind(n)

for line in lines:
    cur = line.split()
    qf.union(int(cur[0]), int(cur[1]))
    
end = time.time()
print("count:", qf.count())
print("time:", int((end - start)*1000), "ms")
print(qf.id)
```

# Quick Union Timed

```python
start = time.time()

qu = QuickUnion(n)

for line in lines:
    cur = line.split()
    qu.union(int(cur[0]), int(cur[1]))
        
end = time.time()
print("count:", qu.count)
print("time:", int((end - start)*1000), "ms")
print(qu.id)
```

# Weighted Quick Union Timed

```python
start = time.time()

wqu = WeightedQuickUnion(n)

for line in lines:
    cur = line.split()
    wqu.union(int(cur[0]), int(cur[1]))
        
end = time.time()
print("count:", wqu.count)
print("time:", int((end - start)*1000), "ms")
print(wqu.id)
```
